```python
import pandas as pd
import matplotlib.pyplot as plt

from IPython.display import set_matplotlib_formats
from matplotlib_inline.backend_inline import set_matplotlib_formats as set_matplotlib_formats_new
```

```python
deform =    pd.read_csv('randbig_deformcnn_ee_kine/progress.csv')
lstm =      pd.read_csv('randbig_lstm_ee_kine/progress.csv')
lstm_nk =   pd.read_csv('randbig_lstm_ee/progress.csv')

fig, axs = plt.subplots(2,1, figsize = (8,8))

axs[0].plot(deform['time/total_timesteps']/1000, deform['rollout/success_rate'], label = 'DeformableCNN Kine', color='green')
axs[0].set_xlabel('Timesteps ($10^3$)', fontsize="12")
axs[0].set_ylabel('Success Rate', fontsize="12")
axs[0].legend(fontsize="12")

axs[0].plot(lstm['time/total_timesteps']/1000, lstm['rollout/success_rate'], label = 'NatureCNN LSTM Kine', color='red')
axs[0].set_xlabel('Timesteps ($10^3$)', fontsize="12")
axs[0].set_ylabel('Success Rate', fontsize="12")
axs[0].legend(fontsize="12")

axs[0].plot(lstm_nk['time/total_timesteps']/1000, lstm_nk['rollout/success_rate'], label = 'NatureCNN LSTM', color='purple')
axs[0].set_xlabel('Timesteps ($10^3$)', fontsize="12")
axs[0].set_ylabel('Success Rate', fontsize="12")
axs[0].legend(fontsize="12")

#######################################################################################################

axs[1].plot(deform['time/total_timesteps']/1000, deform['rollout/ep_rew_mean'], label = 'DeformableCNN Kine', color='green')
axs[1].set_xlabel('Timesteps ($10^3$)', fontsize="12")
axs[1].set_ylabel('Mean Episode Reward', fontsize="12")
axs[1].legend(fontsize="12")

axs[1].plot(lstm['time/total_timesteps']/1000, lstm['rollout/ep_rew_mean'], label = 'NatureCNN LSTM Kine', color='red')
axs[1].set_xlabel('Timesteps ($10^3$)', fontsize="12")
axs[1].set_ylabel('Mean Episode Reward', fontsize="12")
axs[1].legend(fontsize="12")

axs[1].plot(lstm_nk['time/total_timesteps']/1000, lstm_nk['rollout/ep_rew_mean'], label = 'NatureCNN LSTM', color='purple')
axs[1].set_xlabel('Timesteps ($10^3$)', fontsize="12")
axs[1].set_ylabel('Mean Episode Reward', fontsize="12")
axs[1].legend(fontsize="12")


 # Add annotation to top left corner of each subplot
axs[0].text(-0.08, 1, f'({chr(97 + 0)})', transform=axs[0].transAxes, verticalalignment='top', fontsize=12)
axs[1].text(-0.08, 1, f'({chr(97 + 1)})', transform=axs[1].transAxes, verticalalignment='top', fontsize=12)
# Adjust layout

# fig.text(0, 0.5, '', ha='center', va='center', rotation='horizontal', fontsize=12)

plt.tight_layout()
# fig.suptitle('Training Result for LSTM encoders w/wo Robot Kinematics Information')
plt.savefig('sac_lstm_compare_vertical.png', format='png')
```
