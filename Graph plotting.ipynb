```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

from IPython.display import set_matplotlib_formats
from matplotlib_inline.backend_inline import set_matplotlib_formats as set_matplotlib_formats_new
```

```python
mask_no_jitter = pd.read_csv('mask_no_jitter.csv')
only_jitter = pd.read_csv('only_jitter.csv')
only_downsample = pd.read_csv('only_downsample.csv')
```

```python
set_matplotlib_formats_new('svg')
```

```python
dual = pd.read_csv('dual.csv')
single = pd.read_csv('single.csv')

fig, axs = plt.subplots(1,2, figsize = (12,4))
#plt.figure(figsize=(20,4))

# axs[0].plot(mask_no_jitter['time/total_timesteps']/1000, mask_no_jitter['rollout/success_rate'], label = 'SAC Dual Downsampling and Mask')
# axs[0].set_xlabel('Timesteps ($10^3$)')
# axs[0].set_ylabel('Success Rate')
# axs[0].legend(fontsize="8")

axs[0].plot(dual['time/total_timesteps'], dual['rollout/success_rate'], label = 'DDPG Dual')
axs[0].set_xlabel('Timesteps')
axs[0].set_ylabel('Success Rate')
axs[0].legend()

axs[0].plot(single['time/total_timesteps'], single['rollout/success_rate'], label = 'DDPG Single')
axs[0].set_xlabel('Timesteps')
axs[0].set_ylabel('Success Rate')
axs[0].legend(fontsize="8")

# axs[1].plot(mask_no_jitter['time/total_timesteps']/1000, mask_no_jitter['rollout/ep_rew_mean'], label = 'SAC Dual Downsampling and Mask')
# axs[1].set_xlabel('Timesteps ($10^3$)')
# axs[1].set_ylabel('Mean Episode Reward')
# axs[1].legend(fontsize="8")

axs[1].plot(dual['time/total_timesteps'], dual['rollout/ep_rew_mean'], label = 'DDPG Dual')
axs[1].set_xlabel('Timesteps')
axs[1].set_ylabel('Mean Episode Reward')
axs[1].legend()

axs[1].plot(single['time/total_timesteps'], single['rollout/ep_rew_mean'], label = 'DDPG Single')
axs[1].set_xlabel('Timesteps')
axs[1].set_ylabel('Mean Episode Reward')
axs[1].legend()

plt.savefig('sac_plots.svg', format='svg')
```
