```python
import pandas as pd
import matplotlib.pyplot as plt

from IPython.display import set_matplotlib_formats
from matplotlib_inline.backend_inline import set_matplotlib_formats as set_matplotlib_formats_new
```

```python
set_matplotlib_formats_new('svg')
```

```python
deform =    pd.read_csv('randbig_deformcnn_ee_kine/progress.csv')
dual =      pd.read_csv('randbig_dualcnn_ee_kine/progress.csv')
nature =    pd.read_csv('randbig_naturecnn_ee_kine/progress.csv')

fig, axs = plt.subplots(1,2, figsize = (16,4))

axs[0].plot(deform['time/total_timesteps']/1000, deform['rollout/success_rate'], label = 'SAC DeformableCNN')
axs[0].set_xlabel('Timesteps ($10^3$)')
axs[0].set_ylabel('Success Rate')
axs[0].legend(fontsize="8")

axs[0].plot(dual['time/total_timesteps']/1000, dual['rollout/success_rate'], label = 'SAC DualCNN')
axs[0].set_xlabel('Timesteps ($10^3$)')
axs[0].set_ylabel('Success Rate')
axs[0].legend(fontsize="8")

axs[0].plot(nature['time/total_timesteps']/1000, nature['rollout/success_rate'], label = 'SAC NatureCNN')
axs[0].set_xlabel('Timesteps ($10^3$)')
axs[0].set_ylabel('Success Rate')
axs[0].legend(fontsize="8")

#######################################################################################################

axs[1].plot(deform['time/total_timesteps']/1000, deform['rollout/ep_rew_mean'], label = 'SAC DeformableCNN')
axs[1].set_xlabel('Timesteps ($10^3$)')
axs[1].set_ylabel('Mean Episode Reward')
axs[1].legend(fontsize="8")

axs[1].plot(dual['time/total_timesteps']/1000, dual['rollout/ep_rew_mean'], label = 'SAC DualCNN')
axs[1].set_xlabel('Timesteps ($10^3$)')
axs[1].set_ylabel('Mean Episode Reward')
axs[1].legend(fontsize="8")

axs[1].plot(nature['time/total_timesteps']/1000, nature['rollout/ep_rew_mean'], label = 'SAC NatureCNN')
axs[1].set_xlabel('Timesteps ($10^3$)')
axs[1].set_ylabel('Mean Episode Reward')
axs[1].legend(fontsize="8")

plt.savefig('sac_cnn_compare.svg', format='svg')
```
